```python
# UNITY CATALOG — TABLE COMMENTS


spark.sql("""
    COMMENT ON TABLE ecommerce_dev.gold.fact_orders IS 
    'Central fact table for e-commerce analytics. Contains order-level measures 
     joined with payments and reviews. 113,425 rows.'
""")

spark.sql("""
    COMMENT ON TABLE ecommerce_dev.gold.dim_customer IS 
    'Customer dimension table. Contains unique customer details by state and city. 
     99,441 rows.'
""")

spark.sql("""
    COMMENT ON TABLE ecommerce_dev.gold.dim_product IS 
    'Product dimension table. Contains product category and physical dimensions. 
     32,951 rows.'
""")

spark.sql("""
    COMMENT ON TABLE ecommerce_dev.gold.dim_seller IS 
    'Seller dimension table. Contains seller location by state and city. 
     3,095 rows.'
""")

spark.sql("""
    COMMENT ON TABLE ecommerce_dev.gold.dim_date IS 
    'Date dimension table. Contains year, month, day, quarter breakdowns. 
     634 unique dates.'
""")

print("All table comments added.")
```

```python
# OPTIMIZE + ZORDER — fact_orders


spark.sql("""
    OPTIMIZE ecommerce_dev.gold.fact_orders
    ZORDER BY (order_date, customer_id)
""")

print("OPTIMIZE + ZORDER complete.")
```

```python
# DELTA TIME TRAVEL DEMO


# Show full history of fact_orders
print("=== TABLE HISTORY ===")
spark.sql("DESCRIBE HISTORY ecommerce_dev.gold.fact_orders") \
    .select("version", "timestamp", "operation", "operationMetrics") \
    .show(truncate=False)

# Query version 0 (first ever write)
print("=== VERSION 0 ROW COUNT ===")
df_v0 = spark.read.format("delta") \
    .option("versionAsOf", 0) \
    .table("ecommerce_dev.gold.fact_orders")
print(f"Version 0 rows: {df_v0.count()}")

# Query current version
print("=== CURRENT VERSION ROW COUNT ===")
df_current = spark.table("ecommerce_dev.gold.fact_orders")
print(f"Current rows: {df_current.count()}")
```
